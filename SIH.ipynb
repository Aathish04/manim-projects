```python
%%manim -qh --disable_caching -v WARNING SquareToCircle
config.background_color = WHITE

class SquareToCircle(Scene):
   def construct(self):
      StartCircle = VGroup(
         Circle(color=ORANGE,fill_opacity=1),
         Paragraph("VICTIM OR ANYONE","CLOSE BY RAISES AN","SOS REQUEST",color=WHITE,alignment="center").scale(0.25)
      )
      self.play(Write(StartCircle[0]))
      self.play(Write(StartCircle[1]))
      RequestSentRect = VGroup(
         RoundedRectangle(color=PURPLE,fill_opacity=1),
         Paragraph(
            "REQUEST SENT TO SET OF AGENCIES",
            "CHOSEN USING A RECOMMENDATION",
            "SYSTEM BASED ON CAPABILITY, ACTIVITY",
            "AND DISTANCE FROM EMERGENCY",
            color=WHITE,alignment="center").scale(0.25)
      )
      self.play(StartCircle.animate.shift(3*UP).scale(0.7))
      self.play(Write(RequestSentRect[0]))
      self.play(Write(RequestSentRect[1]))
      StartToRequestArrow = Arrow(color=BLACK,stroke_width=3)
      StartToRequestArrowUpdater = lambda arrow: arrow.put_start_and_end_on(StartCircle.get_bottom(), RequestSentRect.get_top())
      StartToRequestArrow.add_updater(StartToRequestArrowUpdater)
      self.play(Write(StartToRequestArrow))
      self.play(RequestSentRect.animate.scale(0.7))
      self.play(VGroup(StartCircle,RequestSentRect).animate.shift(LEFT*4))

      AcceptanceRect = VGroup(
            RoundedRectangle(color=TEAL,width=3,height=2,corner_radius=0.2,fill_opacity=1),
            Paragraph("AGENCY 'A' ACCEPTS","THE REQUEST",alignment="center").scale(0.25))
      self.play(Write(AcceptanceRect[0]))
      self.play(Write(AcceptanceRect[1]))

      RequestToAcceptanceArrow = Arrow(color=BLACK,stroke_width=3)
      RequestToAcceptanceArrowUpdater = lambda arrow: arrow.put_start_and_end_on(RequestSentRect.get_right(),AcceptanceRect.get_left())
      RequestToAcceptanceArrow.add_updater(RequestToAcceptanceArrowUpdater)
      self.play(Write(RequestToAcceptanceArrow))
      self.play(AcceptanceRect[0].animate.scale(0.7))
      self.play(AcceptanceRect.animate.shift(LEFT))

      self.play(VGroup(StartCircle,RequestSentRect,AcceptanceRect).animate.shift(LEFT))
      
      ResourcesAvailable = VGroup(
            Square(color=DARK_BLUE,fill_opacity=1).rotate(45*DEGREES),
            Paragraph("DOES","AGENCY 'A' HAVE THE","NECESSARY RESOURCES","AND PERSONNEL TO", "SERVICE THE REQUEST?",alignment="center").scale(0.24)).shift(RIGHT)
      
      self.play(Write(ResourcesAvailable[0]))
      self.play(Write(ResourcesAvailable[1]))
      
      AcceptToDecisionArrow = Arrow(color=BLACK,stroke_width=3)
      AcceptToDecisionArrowUpdater = lambda arrow: arrow.put_start_and_end_on(AcceptanceRect.get_right(), ResourcesAvailable.get_left())
      AcceptToDecisionArrow.add_updater(AcceptToDecisionArrowUpdater)
      self.play(Write(AcceptToDecisionArrow))
      self.play(ResourcesAvailable.animate.scale(0.75))
      self.play(VGroup(RequestSentRect,AcceptanceRect,ResourcesAvailable).animate.shift(UP))
      
      YesCircle = VGroup(Circle(radius = 0.25,fill_opacity=1,color=GREEN_E),Text("YES").scale(0.25)).shift(RIGHT+DOWN*1)
      YesCircle.set_opacity(0)
      self.add(YesCircle)
      DecisionToYesArrow = Arrow(color=BLACK,stroke_width=3)
      DecisionToYesArrow.put_start_and_end_on(ResourcesAvailable.get_bottom(), YesCircle.get_top())
      DecisionToYesArrowUpdater = lambda arrow: arrow.put_start_and_end_on(ResourcesAvailable.get_bottom(), YesCircle.get_top())
      DecisionToYesArrow.add_updater(DecisionToYesArrowUpdater)
      self.wait(1)
      self.play(Write(VGroup(DecisionToYesArrow)))
      # YesCircle.set_opacity(1)
      self.play(YesCircle.animate.set_opacity(1))
      Final = VGroup(Circle(color=GREEN,fill_opacity=1),Paragraph("AGENCIES", "COLLABORATE AND","SUCCESSFULLY","SERVICE THE","EMERGENCY!",alignment="center").scale(0.25))
      Final.shift(LEFT*4+DOWN*1.5)
      self.wait(1)
      self.play(Write(Final[0]))
      self.play(Write(Final[1]))

      YesToFinalArrow = Arrow(color=BLACK,stroke_width=3)
      YesToFinalArrow.put_start_and_end_on(YesCircle.get_left(), Final.get_right())
      YesToFinalArrowUpdater = lambda arrow: arrow.put_start_and_end_on(YesCircle.get_left(), Final.get_right())
      YesToFinalArrow.add_updater(YesToFinalArrowUpdater)
      self.wait(1)
      self.play(Write(YesToFinalArrow))


      NoCircle = VGroup(Circle(radius = 0.25,fill_opacity=1,color=RED),Text("NO").scale(0.25)).move_to(ResourcesAvailable.get_right()+RIGHT*2)
      DecisionToNoArrow = Arrow(color=BLACK,stroke_width=3)
      DecisionToNoArrowUpdater = lambda arrow: arrow.put_start_and_end_on(ResourcesAvailable.get_right(), NoCircle.get_left())
      DecisionToNoArrow.add_updater(DecisionToNoArrowUpdater)
      self.play(Write(VGroup(NoCircle)))
      self.play(Write(VGroup(DecisionToNoArrow)))
      self.wait(1)
      RaiseRequestCircle = VGroup(
         Circle(color=ORANGE,fill_opacity=1),
         Paragraph("AGENCY A","RAISES REQUEST","FOR RESOURCES/","PERSONNEL","REQUIRED TO","SERVICE","THE REQUEST",alignment="center").scale(0.25)
      ).move_to(NoCircle.get_center()+DOWN*2)
      NoToRaiseRequestArrow = Arrow(color=BLACK,stroke_width=3)
      NoToRaiseRequestArrow.put_start_and_end_on(NoCircle.get_bottom(), RaiseRequestCircle.get_top())
      NoToRaiseRequestArrowUpdater = lambda arrow: arrow.put_start_and_end_on(NoCircle.get_bottom(), RaiseRequestCircle.get_top())
      NoToRaiseRequestArrow.add_updater(NoToRaiseRequestArrowUpdater)
      self.wait(1)
      
      self.play(Write(RaiseRequestCircle))
      self.play(Write(NoToRaiseRequestArrow))
      self.wait(1)
      A2ARequest = VGroup(
         RoundedRectangle(height=1.5,width=3.0,color=PURPLE_B,fill_opacity=1),
         Paragraph(
            "This request is sent".upper(),
            "to a set of agencies".upper(),
            "using a recommendation".upper(),
            "system based on".upper(),
            "the resources the".upper(),
            "agency owns,".upper(),
            "its activity status".upper(),
            "and its location.".upper(),color=WHITE,alignment="center").scale(0.25)).scale(0.8)
      A2ARequest.shift(RaiseRequestCircle.get_bottom()+DOWN*1.1)




      self.play(Write(A2ARequest))
      Req2ReqSentArrow = Arrow(color=BLACK,stroke_width=3)
      Req2ReqSentArrowUpdater = lambda arrow: arrow.put_start_and_end_on(RaiseRequestCircle.get_bottom(), A2ARequest.get_top())
      Req2ReqSentArrow.add_updater(Req2ReqSentArrowUpdater)
      self.wait(1)
      self.play(Write(Req2ReqSentArrow))
      Dots = VGroup(Dot(color=BLACK),Dot(color=BLACK),Dot(color=BLACK),Dot(color=BLACK),Dot(color=BLACK),Dot(color=BLACK),Dot(color=BLACK),Dot(color=BLACK))
      # Dots.move_to()
      Computers =VGroup(
            SVGMobject("/home/aathishs/Desktop/manim/monitor-icon.svg",fill_color=BLUE).shift(UP*3.5),
            SVGMobject("/home/aathishs/Desktop/manim/monitor-icon.svg",fill_color=GREEN).shift(DOWN*3.5),
            SVGMobject("/home/aathishs/Desktop/manim/monitor-icon.svg",fill_color=RED).shift(RIGHT*3.5),
            SVGMobject("/home/aathishs/Desktop/manim/monitor-icon.svg",fill_color=DARK_BLUE).shift(LEFT*3.5),
            SVGMobject("/home/aathishs/Desktop/manim/monitor-icon.svg",fill_color=PINK).shift(DL*3.5),
            SVGMobject("/home/aathishs/Desktop/manim/monitor-icon.svg",fill_color=PURPLE).shift(UR*3.5),
            SVGMobject("/home/aathishs/Desktop/manim/monitor-icon.svg",fill_color=ORANGE).shift(DR*3.5),
            SVGMobject("/home/aathishs/Desktop/manim/monitor-icon.svg",fill_color=LIGHT_GRAY).shift(UL*3.5),
            SVGMobject("/home/aathishs/Desktop/manim/monitor-icon.svg")
            ).scale(0.2).to_corner(UR)
      Dots.move_to(Computers[-1])
      Targets = Computers[:-1]
      self.play(FadeIn(Computers),FadeIn(Dots))
      # self.add(Dots)
      sendrequestsanim = []
      for i,target in enumerate(Targets):
         sendrequestsanim.append(Dots[i].animate.move_to(target.get_center()))
      self.play(LaggedStart(*sendrequestsanim,lag_ratio=0.2))
      self.play(Dots[-1].animate.set_color(GREEN_E))
      self.wait(1)
      self.play(FadeOut(Targets[:-1],Dots[:-1]))
      self.play(Dots[-1].animate.move_to(Computers[-1]))
      self.wait(1)
      self.play(FadeOut(Dots[-1]))
      self.play(FadeOut(Computers[-2:]))





      AcceptanceBRect = VGroup(
                  RoundedRectangle(color=TEAL_B,width=2,height=1,corner_radius=0.2,fill_opacity=1),
                  Paragraph(
                     "AGENCY 'B' ACCEPTS","WITH SUFFICIENT","RESOURCES ACCEPTS","THE REQUEST",
                     alignment="center").scale(0.25)).move_to(A2ARequest.get_left()+LEFT*2)
      self.play(Write(AcceptanceBRect[0]))
      self.play(Write(AcceptanceBRect[1]))
      self.wait(1)
      
      RequestToBAcceptArrow = Arrow(color=BLACK,stroke_width=3)
      RequestToBAcceptArrow.put_start_and_end_on(NoCircle.get_bottom(), RaiseRequestCircle.get_top())
      RequestToBAcceptArrowUpdater = lambda arrow: arrow.put_start_and_end_on(A2ARequest.get_left(), AcceptanceBRect.get_right())
      RequestToBAcceptArrow.add_updater(RequestToBAcceptArrowUpdater)
      self.wait(1)
      self.play(Write(RequestToBAcceptArrow))
      self.wait(1)
      
      AcceptanceToCollabArrow = Arrow(color=BLACK,stroke_width=3)
      AcceptanceToCollabArrow.put_start_and_end_on(AcceptanceBRect.get_left(), Final.get_right())
      AcceptanceToCollabArrowUpdater = lambda arrow: arrow.put_start_and_end_on(AcceptanceBRect.get_left(), Final.get_right())
      AcceptanceToCollabArrow.add_updater(AcceptanceToCollabArrowUpdater)
      
      self.play(Write(AcceptanceToCollabArrow))
      self.wait(1)
```
